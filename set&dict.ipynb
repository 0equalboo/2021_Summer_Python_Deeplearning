# **세트(set)**
* 세트는 순서가 보장되지 않음
* 세트는 요소의 중복을 허용하지 않음
* 세트는 {}를 활용하여 생성이 가능 

### **생성**



```python
myset = set()
myset
```

```python
myset = set([1,1,1,2,2,2,3,3,3])
myset
```

```python
myset = set((1,2,3,4,5,6,6,6,))
myset
```

### **add : 값 추가**
* set에 요소를 추가

빈 set 생성

```python
myset = set()
myset
```

add() 로 요소 추가

```python
myset.add(1)
myset
```

```python
myset.add(1)
myset.add(1)
myset.add(2)
myset.add(2)
myset.add(2)
myset.add(3)
myset.add(3)
myset.add(4)
myset.add(5)
myset
```

요소를 중복하여 추가하여도, set의 특성상 **중복이 제거된 요소만 출력**

## **update() : 여러개의 값 추가**
* 여러 개의 값을 한꺼번에 추가하고자 할때 사용하는 메서드 

```python
myset = {1,2,3}
myset
```

list의 요소들을 set에 한꺼번에 **update**

```python
myset.update([4,5,6])
myset
```

tuple의 요소들을 set에 한꺼번에 **update**

```python
myset.update((7,8,9))
myset
```

set의 요소들을 set에 한꺼번에 **update**

```python
myset.update({10,11,12})
myset
```

## **remove() : 값 제거**
* 단일 요소를 제거

```python
myset = {1,2,3,4,5}
```

```python
myset.remove(2)
myset
```

## **교집합 (intersection)**
* 교집합은 집합 a와 집합 b의 공통된 요소
* & 기호 또는 intersection() 메서드를 활용해 구할 수 있음

```python
a = {1,2,3,4,5}
b = {3,4,5,6,7}
a & b
```

```python
a.intersection(b)
```

## **합집합 (union)**
* 합집합은 집합 a와 집합 b 요소를 모두 포함
* | 기호 또 union() 매서드를 사용

## **차집합 (difference)**
* 두 집합에서 하나의 집합에만 포함되는 모든 원소의 집합
* -연산자 또는 differnce() 매서드를 사용 

```python
a = {1,2,3,4,5,6,7}
b = {5,6,7,8,9,10,11}
a-b
```

```python
b-a
```

```python
a.difference(b)
```

# **딕셔너리(dictionary)**
* 순서 x
* 키(key)와 값(value)의 쌍으로 형성
* type = dict으로 표기
* key를 사용하여 값 조회
* **수정, 삭제, 추가 가능** 

**생성**

```python
mydict = dict()
print(type(mydict))
mydict
```

```python
mydict = {'a' : 1, 'b':2, 'c':3}
mydict
```

딕셔너리는 여러 타입의 key 가질 수 있음

```python
mydict = {'a':1,'가':2,100:3,3.14:4,True:5}
mydict
```

**값 조회**
* key 값을 통해 값을 조회할 수 있음

key를 지정하여 값을 조회할 수 있음

```python
mydict['a']
```

없는 key를 조회할 경우 오류 발생

```python
mydict['b']
```

get() 메서드를 활용한 값 조회 가능

```python
mydict.get('a')
```

get() 사용시 key가 없는 경우에 오류가 발생하지 않음
* 이 경우에는 None값 반환

```python
mydict.get('b')
```

### **keys() : 모든 key 조회**

```python
mydict = {'a': 100, 'b': 200, 'c': 300, 'd':400, 'e':500}
mydict
```

```python
mydict.keys()
```

**주의사항** dict_keys
* 위에서 출력된 dict_keys는 리스트(list)가 아님
* 객체(object)로 생성되는데, 이를 list로 변경하기 위해서 list()로 타입변환 가능

```python
a = mydict.keys()
b = ['a','b','c','d','e']
a
```

```python
type(a)
```

```python
type(b)
```

```python
a = list(a)
type(a)
```

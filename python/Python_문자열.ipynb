## **문자열 출력 (포맷팅)**

**print() : 출력**

```python
print('Hello pyhon')
```

```python
print('Hello','Python')
```

end 옵션으로 출력될 마지막 글자 변경 가능(default : \n)
* [참고] : \n 은 개행을 의미하는 특수문자

```python
print('여기까지', end='\n')
print('끝')
```

```python
print('여기까지\n')
print('끝')
```

sep 옵션으로 출력될 문자열 사이에 출력될 글자를 변경할 수 있음(default : 공백)

```python
print('여기','까지','끝',sep='==')
```

**%를 사용한 출력**

```python
"Hello %s" % ('world')
```

```python
'Hello %.3f' % (0.1234)
```

```python
'Hello %d' % (12345)
```

```python
'Hello %c' % ('A')
```

**{}와 format를 사용한 출력**

```python
'Welcome to {}'.format('python')
```

```python
'password {}'.format(4885)
```

소수점 2자리까지 제한하는 경우

```python
'원주율 {:.2f}'.format(3.141592)
```

**f 문자열 포맷팅 (python 3.6 이상만 지원)**

```python
name = 'Boo'
age = 17
print(f'나의 이름은 {name}입니다. 나이는 {age} 살입니다.')
```

```python
print(f'내년에 저는 {age+1} 살입니다')
```

```python
d = {'name':'Boo','age':17}
print(f"반가워요.저는 {d['name']}입니다. 저의 나이는 {d['age']} 살입니다.")
```

## **인덱싱 (indexing)**
* 문자열에서 한 개의 글자를 조회하기 위해서 **[]를 활용한 인덱싱으로 조회 가능**

```python
a = 'Python is my life'
a
```

```python
a[0]
```

```python
a[-2]
```

## **슬라이싱 (Slicing)**
* 슬라이싱의 활용 : [start:stop:step] 을 명시하여 부분 추출 가능

[:]을 활용해 전체 추출

```python
a[:]
```

[start:]는 시작 index부터 끝까지 추출

```python
a[3:]
```

```python
a[-4:]
```

[:end]는 처음부터 **end 전까지** 추출

```python
a[:6]
```

```python
a[:-3]
```

[start:end]는 start부터 end 전까지 추출

```python
a[3:6]
```

[start:stop:step]에서 step 지정시 step 만큼 건너뛰며 추출

```python
a[::2]
```

역순으로 출력하기 위에서는 step을 -1로 지정

```python
a[::-1]
```

### **불변(immutable)**
* 문자열은 index를 통해 값을 수정 및 변경이 불가능
* 하지만, 메서드(method)를 통해 변경 가능

```python
a = 'Python is my life'
```

index에 값을 할당하여 변경하려는 경우 오류 발생

```python
a[2] = 'Y'
```

del로 부분 삭제하려는 경우 오류 발생

```python
del a[2]
```

## **문자열(텍스트)을 다루는 다양한 기능, 메서드(method)**
문자열만 가지고 있는 **고유의 기능**

### **split() : 분리**
* split은 문장을 특정 규칙에 의해 쪼개 주는 기능
* 분리한 결과는 list형식으로 값을 return

```python
a = 'This is a pen'
```

split()의 기본값은 공백이며, 공백을 기준으로 분리하여 리스트로 반환

```python
a.split()
```

기본 값인 공백에서 **특정 문자**로 지정 가능

```python
a = 'This-is-a-pen'
a.split('-')
```

aa에서 반환된 리스트를 대입 후 index로 값 조회 가능

```python
aa = a.split('-')
aa[0]
```

```python
aa[0] + ' ' + aa[2]
```

### **join(): 합치기**
* 결합하고자 하는 문자에 .join() 안에 리스트를 지정하여 결합 가능

```python
a = '010 1234 5678'
'-'.join(a.split())
```

```python
'-'.join('ABCDE')
```

### **startswith, endswith**

시작과 끝이 맞는지 결과를 bool로 반환

```python
a = '01-sample.png'
```

**startswith() : 시작하는**
* 지정한 문자열로 시작하면 True, 그렇지 않으면 False를 반환

```python
a.startswith('01')
```

```python
a.startswith('01-sample.png')
```

```python
a.startswith('02')
```

**endswith() : 끝나는**
* 지정한 문자열로 끝나면 True, 그렇지 않다면 False를 반환

```python
a.endswith('png')
```

```python
a.endswith('jpg')
```

## **replace() : 문자열 바꾸기**
* 문자열에서 replace(바꿀대상, 바꾸려는 문자열)를 지정하여 문자열 변경
* 결과는 복사본이 만들어져 반환

```python
a = '01-sample.png'
```

```python
a.replace('01','02')
```

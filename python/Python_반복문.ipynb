# **반복문**
* 1회 이상 반복해서 수행하는 일을 반복문으로 쉽게 해결 가능
* list, tuple, dict, set 등 집합에 대한 순회를 돌며 일을 처리할 때 많이 사용
* list, tuple, dict, set, 문자열까지 모두 순회가능한 객체
* 순회가능한 객체는 반복문을 통해 순회 가능

```python
mylist = [1,2,3,4,5,6,7]
```

## **for 와 in 구문 (반복문)**
**[기본 문법]**
for (+) 하나씩 꺼내올 때 변수 (+) in (+) [꺼내올 집합]:
* list, tuple, set, dictionary, 문자열 형태 모두 가능
* range와 결합해 사용 가능

### **리스트(list)**

```python
for i in mylist:
  print(i)
```

### **튜플(tuple)**

```python
for i in (1,2,3,4,5):
  print(i)
```

### **tuple + list**

tuple을 전체로 받아주는 경우

```python
person = ('Jeck', 10)
print(person)
print(person[0])
print(person[1])
```

tuple의 요소를 개별로 받아주는 경우

```python
name, age = ('Bam',10)
print(name)
print(age)
```

반복문에서 응용방법

```python
mytuplelist = [('Jeck',10),('Peter',20),('Tiger',30)]
for mytuple in mytuplelist:
  print(mytuple[0], mytuple[1])
```

```python
mytuplelist = [('Jeck',10),('Peter',20),('Tiger',30)]
for name, age in mytuplelist:
  print(name, age)
```

### **딕셔너리(dicitionary)**

```python
mydict = {'Con':50, 'Iron':60,'Peng':70}
for key in mydict.keys():
  print(key)
```

```python
for value in mydict.values():
  print(value)
```

```python
for name, age in mydict.items():
  print(name, age)
```

### **문자열(str)**

```python
for c in "Hello":
  print(c)
```

### **range()**
* range() 함수는 별도의 list, tuple 생성 없이 range() 에서 정의한 범위를 반복하는데 활용할 수 있음
* range(start, stop, step) 형식을 사용


stop : 단일 값을 지정하는 경우

```python
for i in range(10):
  print(i)
```

start, stop : 두개의 값을 지정한 경우

```python
for i in range(2,9):
  print(i)
```

start, stop, step : 세 개의 값을 지정한 경우

```python
for i in range(1, 10, 2):
  print(i)
```

### **반복문의 중첩**

```python
for i in range(1,4):
  for j in range(1,4):
    print(f'(i={i}) X (j={j}) = {i * j}')
  print('===')
```

## **제어문**
**continue**
* 반복문 내부에서 continue 구문은 해당 루프(loop)를 건너뛰는 역할을 수행
* continue라는 구문을 만나면, 반복문에서 continue 아래 작성된 코드는 실행되지 않고 건너뜀

```python
mylist = [1,2,3,4,5,6,7,8,9]
```

**반복문**과 **조건문** 그리고 continue를 활용하여 **짝수**만 출력이 가능

```python
for i in mylist:
  if i % 2 == 1:
    continue
  print(i)
```

### **break**
* **break** 구문을 만나면 반복 루프(loop)가 즉시 종료됨

```python
mylist = [1,2,3,4,5,6,7,8]
```

break를 사용하여 i가 6 이상일시 stop하는 코드 작성

```python
for i in mylist:
  if i >= 6:
    break
  print(i)
```

### **break, continue 차이**

```python
for i in range(10):
  if i ==5:
    break
  print(i)
```

```python
for i in range(10):
  if i == 5:
    continue
  print(i)
```

## **while**
* while문은 while문과 함께 정의한 조건이 참인 동안만 반복 루프 수행

```python
count = 5 
while count > 0:
  print(count)
  count -=1
```

while True로 지정후 무한루프 생성

```python
count = 1
while True:
  print(count)
  count += 1
  if count > 5:
    break
```

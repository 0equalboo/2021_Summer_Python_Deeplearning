## **조건문**
주어진 조건이 참인 경우그 다음내가 정한 규칙(logic)을 실행하는개념

**if**
* if는 어떤조건이 성립한다면 ~이라는 뜻을 가짐
* if구문끝에는 반드시 콜론(:) 필요
* if구문에 로직을 포함시키기 위해선 들여쓰기로 구분하여야 함

```python
if 5>3:
  print('참')
```

```python
if 5<3:
  print('참')
  print('참')
  print('참')

print('끝')
```

```python
if 5<3:
  print('참')
```

**else**  
else는 if 조건 후에 따라오며, if가 아닌 경우에 실행됨

```python
if 5<3:
  print('if 구문 실행')
else:
  print('else 구문 실행')
```

else 구문을 단독으로 실행할 시 오류 발생

```python
else:
  print('else 구문 실행')
```

**elif**  
elif 구문은 3가지 이상 분기 (조건)의 동작을 수행할 때 사용

```python
if 3>5:
  print('if 구문')
elif 3<4:
  print('elif 구문')
else:
  print('None')
```

```python
if 3>5:
  print('if 구문')
elif 3>4:
  print('elif 1구문')
elif 4>3:
  print('elif 2구문')
elif 3>1:
  print('elif 3구문')
else:
  print('None')
```

if문은 위에서부터 차례대로 실행

# **1이나 0이 참 또는 거짓을 표현하는 경우**

```python
if 1:
  print('참')
else:
  print('거짓')
```

```python
if 0:
  print('참')
else:
  print('거짓')
```

## **삼항연산자**
(문법)**(참인 값) if 조건 else(거짓인 경우 값)**


```python
age = 35
"30세 이상입니다." if age >= 30 else "30세 이하입니다."
```

```python
age = 20
"30세 이상입니다." if age >=30 else '30세 이하입니다'
```

**and** / **or** : 두가지 이상 조건을 다룰 때 활용

**and**  
* and조건은 모두 만족할 때 참으로 인식

```python
True and True and True
```

```python
True and False and True
```

```python
if (0<1) and (0<2):
  print('모두 참')
else:
  print('거짓 포함')
```

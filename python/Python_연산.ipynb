# **연산**
Excel에서도 sum이나 +,- 기호로 셀과 셀 단위, 열과 열 단위, 행과 행단위 연산이 가능

## **사칙연산(+,-,*,/)**
* 더하기(+)
* 빼기(-)
* 곱하기(*)
* 나누기(/)

```python
a = 10
b = 3
```

**+ : 덧셈**

```python
a+b
```

```python
10+234
```

+= 기호는 +(덧셈) 이후 자기 자신의 변수에 재할당  
ex) a+=3 은 a = a+3 과 같음

```python
s = 10
s += 5
print(s)
```

**- : 뺄셈**

```python
a - b
```

```python
10  - 9
```

```python
s = 10
s-=3
print(s)
```

**/ : 나눗셈**

```python
a/b
```

```python
10.7/7
```

```python
s = 10
s/=2
print(s)
```

*** : 곱셉**

```python
a*b
```

```python
3*4
```

```python
s = 10
s*=3
print(s)
```

## **기타 연산자(%,//,****)
* // : floor division 연산자. 나눗셈에 대한 몫을 정수형으로 구함
* % : modulus 연산자. 몫을 나눈 후 나머지를 구함
* ** : 제곱 연산 

```python
a = 10
b = 3
a//b
```

```python
a=10
b=3
a%b
```

```python
a=10
b=3
a**b
```
